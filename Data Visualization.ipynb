## Data Visualization

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
%matplotlib inline
```

```python
sns.set()
```

```python
iris=sns.load_dataset("iris")
```

```python
iris.head()
```

```python
iris.info()
```

```python
iris.describe()
```

## scatter plot

```python
plt.plot(iris.sepal_length,iris.sepal_width, "o")
```

```python
plt.plot(iris.sepal_length,iris.sepal_width, "+")
```

```python
plt.scatter(iris.sepal_length,iris.sepal_width, 
            s=np.random.randint(50,200,150),  ### size 
            c=np.random.randint(0,10,150),  ## color , pixel 
            cmap="Reds",
            edgecolor="black",
            linewidth=1, 
            alpha=1)
plt.xlabel("sepal_width")
plt.ylabel("petal_length")
plt.title("Ä°ris Sepal Length & Sepal Width ")
```

```python
iris.plot.area()
```

```python
ax=iris.plot.scatter(x='sepal_length', y='sepal_width', 
                     color='Blue', label='sepal')
iris.plot.scatter(x='petal_length', y='petal_width', color='red', 
                  label='petal', ax=ax)
```

### Pie Chart

```python
iris_avg=iris["petal_width"].groupby(iris["species"]).mean()
iris_avg
```

```python
iris_avg.plot.pie()
```

```python
iris_avg_2=iris[["petal_width", 
                 "petal_length"]].groupby(iris["species"]).mean()
```

```python
iris_avg_2
```

```python
iris_avg_2.plot.pie(subplots=True)
```

```python
iris_avg.plot.pie(labels=["setosa","versicolor", "virginica"], colors=list("brg"), 
           autopct='%.2f', fontsize=25, figsize=(10,10))
```
